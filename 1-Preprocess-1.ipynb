# Dados de Bolsonaro



```python
# datas são lidas como string e depois convertidas
column_types = {'ID' : int, 'DataPost': str, 'DiaDaSemana': str}
```

```python
#df1 = pd.read_excel(base_path + 'bolsonaro-preproc0-manual.xlsx', index_col=0)
df1 = pd.read_excel(base_path + 'Bolsonaro Validado.xlsx', index_col=0)
df1.reset_index(drop=True, inplace=True)
df1.set_index('ID', inplace=True)
df1.head(5)
```

```python
df1.drop(columns=["DataColeta", "LinkFoto", "LinkVideo", "LinkPost"], inplace=True)
df1.head(5)
```

```python
# mostra tamanho da tabela com NA em alguma entrada
df1.loc[df1.isna().sum(axis=1) > 0].shape
```

```python
df1.dropna(inplace=True)
```

```python
df1.info()
```

```python
# converte para datetime 
df1['DataPost'] = pd.to_datetime(df1['DataPost'])
df1['DiaDaSemana'] = pd.to_datetime(df1['DiaDaSemana'])
```

```python
# converte esta coluna para conter uma string com o dia da semana
df1['DiaDaSemana'] = df1['DiaDaSemana'].dt.strftime('%A')
```
