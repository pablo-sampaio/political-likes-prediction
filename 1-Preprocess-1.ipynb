```python
# converte para datetime 
def adjust_dates(df_aux):
    df_aux['DataColeta'] = pd.to_datetime(df_aux['DataColeta'], format='mixed')
    df_aux['DataPost'] = pd.to_datetime(df_aux['DataPost'], format='mixed')

    try:
        # converte esta coluna para conter uma string com o dia da semana
        df_aux['DiaDaSemana'] = pd.to_datetime(df_aux['DiaDaSemana'])
        df_aux['DiaDaSemana'] = df_aux['DiaDaSemana'].dt.strftime('%A')
    except:
        print("Coluna DiaDaSemana parece já estar guardado como date. Não precisa transformar!")
```

```python
#df1 = pd.read_excel(base_path + 'Bolsonaro + Lula - Validado.xlsx') # está com problemas na coluna de duração do vídeo

df_bols = pd.read_excel(base_path + 'Bolsonaro Validado.xlsx', dtype=column_types)
df_lula = pd.read_excel(base_path + 'LulaTotal Validado.xlsx', dtype=column_types)

df_bols.shape, df_lula.shape
```

```python
# concatena os dois dataframes
df1 = pd.concat([df_bols, df_lula], axis=0)
df1.shape
```

```python
# para garantir que os IDs são únicos, mesmo entre diferentes candidatos
len(df1['ID'].unique())
```

```python
df1.reset_index(drop=True, inplace=True)
df1.set_index('ID', inplace=True)
df1.head(5)
```

```python
df1.drop(columns=["LinkFoto", "LinkVideo", "LinkPost"], inplace=True)
df1.head(5)
```

```python
df1.dropna(inplace=True)
```

```python
# converte para datetime
adjust_dates(df1)
df1.head(5)
```

```python
df1['DiasDecorridos'] = (df1['DataColeta'] - df1['DataPost']).dt.days
```
